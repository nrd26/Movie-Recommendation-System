### Merging the datasets
Since it's a lot easier to just work with 1 dataset, I just merged both of them


```python
data1.columns = ['id','tittle','cast','crew']
data= data.merge(data1,on='id')
```

### Analysing the data
Here we use literal_eval to check if our data is a python object or not(ast.literal_eval raises an exception if the input isn't a valid Python datatype, so the code won't be executed if it's not.)

```python
from ast import literal_eval

features = ['cast', 'crew', 'keywords', 'genres']
for feature in features:
    data[feature] = data[feature].apply(literal_eval)
```

A simple function to get the Director's name from the crew list( returns NaN if there is no mention of the director)

```python
def director(x):
    for i in x:
        if i['job'] == 'Director':
            return i['name']
    return np.nan
```

For using the data on cast, movie keywords and genre, we select the top 5(If there are >=5 elements in each category we return the top 5 else we return the entire list and return an empty list in case of missing data)

```python
def top_features(x):
    if isinstance(x, list):
        names = [i['name'] for i in x]
        if len(names) > 5:
            names = names[:5]
        return names

    return []
```

We now extract the data we want from the data we have

```python
data['director'] = data['crew'].apply(director)

features = ['cast', 'keywords', 'genres']
for feature in features:
    data[feature] = data[feature].apply(top_features)
```

```python
# Print the new features
data[['title', 'cast', 'director', 'keywords', 'genres']].head()
```

This is a function to remove whitespaces and convert all strings to lowercase so that Python can differentiate between say for example a Chris Evans and a Chris Hemsworth and counts them as different entities(it returns an empty string if there's no info on the director)

```python
def remove(x):
    if isinstance(x, list):
        return [str.lower(i.replace(" ", "")) for i in x]
    else:
        if isinstance(x, str):
            return str.lower(x.replace(" ", ""))
        else:
            return ''
```

```python
features = ['cast', 'keywords', 'director', 'genres']

for feature in features:
    data[feature] = data[feature].apply(remove)
```

This function creates a string with all the metadata that we want

```python
def result(x):
    return ' '.join(x['keywords']) + ' ' + ' '.join(x['cast']) + ' ' + x['director'] + ' ' + ' '.join(x['genres'])
data['result'] = data.apply(result, axis=1)
```

We now implement the CountVectorizer from sklearn to analyse the words

```python
from sklearn.feature_extraction.text import CountVectorizer

count = CountVectorizer(stop_words='english')
count_matrix = count.fit_transform(data['result'])
```

The below code computes the cosine Similarity among the vectors

```python
from sklearn.metrics.pairwise import cosine_similarity

cos_sim = cosine_similarity(count_matrix, count_matrix)
```

The below code resets the indices

```python
data = data.reset_index()
indices = pd.Series(data.index, index=data['title'])
```

The below function takes in movie title as input and outputs most similar movies 

```python
indices = pd.Series(data.index, index=data['title']).drop_duplicates()

def find_rec(title, cos_sim=cos_sim):
    idx = indices[title]
    sim_scores = list(enumerate(cos_sim[idx]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    sim_scores = sim_scores[1:11]
    movie_indices = [i[0] for i in sim_scores]

    return data['title'].iloc[movie_indices]
```
